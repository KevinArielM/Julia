```julia
n = 100

x = range(-2, 2, length = n)
y = x
z = zeros(n, n)
f(x, y) = sin(x^2 + y^2)/(x^2 + y^2)
for i  1:n
for j  1:n
z[i, j] = f(x[i], y[j])
end
end
plot(x, y, f, st = :surface, xlabel = "x", ylabel = "y")
```
